```python
df_train_windows=df_2017[(df_2017['date']<=pd.datetime(2017,8,15))&(df_2017['date']>=pd.datetime(2017,5,15))]
df_cv=df_2017[(df_2017['date']>=pd.datetime(2017,7,15))&(df_2017['date']<=pd.datetime(2017,7,31))]
df_cv_windows=df_2017[(df_2017['date']<=pd.datetime(2017,7,15))&(df_2017['date']>=pd.datetime(2017,4,15))]
df_train1=df_2017[(df_2017['date']>=pd.datetime(2017,6,15))&(df_2017['date']<=pd.datetime(2017,6,30))]
df_train1_windows=df_2017[(df_2017['date']<=pd.datetime(2017,6,15))&(df_2017['date']>=pd.datetime(2017,3,15))]
df_train2=df_2017[(df_2017['date']>=pd.datetime(2017,5,15))&(df_2017['date']<=pd.datetime(2017,5,30))]
df_train2_windows=df_2017[(df_2017['date']<=pd.datetime(2017,5,15))&(df_2017['date']>=pd.datetime(2017,2,15))]
df_train3=df_2017[(df_2017['date']>=pd.datetime(2017,4,15))&(df_2017['date']<=pd.datetime(2017,4,30))]
df_train3_windows=df_2017[(df_2017['date']<=pd.datetime(2017,4,15))&(df_2017['date']>=pd.datetime(2017,1,15))]
```

```python
df_test_windows_rec=df_train_windows.groupby(['store_nbr','item_nbr']).size().to_frame('records_count0')
df_cv_windows_rec=df_cv_windows.groupby(['store_nbr','item_nbr']).size().to_frame('records_count_cv')
df_train1_windows_rec=df_train1_windows.groupby(['store_nbr','item_nbr']).size().to_frame('records_count1')
df_train2_windows_rec=df_train2_windows.groupby(['store_nbr','item_nbr']).size().to_frame('records_count2')
df_train3_windows_rec=df_train3_windows.groupby(['store_nbr','item_nbr']).size().to_frame('records_count3')


df_cv_rec=df_cv.groupby(['store_nbr','item_nbr']).size().to_frame('records_count_cv')
df_train1_rec=df_train1.groupby(['store_nbr','item_nbr']).size().to_frame('records_count1')
df_train2_rec=df_train2.groupby(['store_nbr','item_nbr']).size().to_frame('records_count2')
df_train3_rec=df_train3.groupby(['store_nbr','item_nbr']).size().to_frame('records_count3')

for i,j in zip([df_cv_windows_rec,df_train1_windows_rec,df_train2_windows_rec,df_train2_windows_rec],[df_cv_rec,df_train1_rec,df_train2_rec,df_train3_rec]):
    print('特征窗口条目：{0}，采样窗口条目：{1}'.format(i.shape,j.shape))
```
